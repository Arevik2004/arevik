# 1.Создайте простой граф, описывающий некоторых ваших близких друзей, ваши отношения с ними, а также некоторые отношения между друзьями. Начните с трёх узлов, включая себя, и создайте пять отношений

```python
CREATE (:Person {name: "Аревик Элбакян", age: 24, city: "Ставрополь"});
CREATE (:Person {name: "Иван Иванов", age: 26, city: "Ставрополь"});
CREATE (:Person {name: "Александра Элбакян", age: 37, city: "Санкт-Петербург"});
MATCH (a:Person {name: "Аревик Элбакян"}), (n:Person {name: "Иван Иванов"}) CREATE (a)-[:ДРУЗЬЯ]->(n);
MATCH (a:Person {name: "Аревик Элбакян"}), (s:Person {name: "Александра Элбакян"}) CREATE (a)-[:ДРУЗЬЯ]->(s);
MATCH (n:Person {name: "Иван Иванов"}), (s:Person {name: "Александра Элбакян"}) CREATE (n)-[:ДРУЗЬЯ]->(s);
MATCH (a:Person {name: "Аревик Элбакян"}), (n:Person {name: "Иван Иванов"}) CREATE (a)-[:ОДНОГРУППНИКИ]->(n);
MATCH (a:Person {name: "Аревик Элбакян"}), (s:Person {name: "Александра Элбакян"}) CREATE (a)-[:РОДСТВЕННИКИ]->(s);
MATCH (n) RETURN n;
```
