# 1. Используйте cURL, чтобы добавить новый документ в базу данных music с определенным _id на ваш выбор.

# 3. CouchDB поддерживает вложения, которые представляют собой произвольные файлы, которые можно сохранять вместе с документами (аналогично вложениям электронной почты). Опять же, используя cURL, создайте новый документ, содержащий текстовый документ в качествевложения. Наконец, составьте и выполните запрос cURL, который вернет только это вложение документа

```python
# 1. Создаем специальную базу для вложений
curl -X PUT http://localhost:5984/attachments_demo
```

```python
#Результат терминала
 {"ok":true}
```

```python
# 2. Создаем документ и сразу получаем его _id и _rev
echo "Создаем основной документ..."
RESPONSE=$(curl -s -X POST http://localhost:5984/attachments_demo \
  -H "Content-Type: application/json" \
  -d '{
    "type": "document_with_text_attachment",
    "name": "Тестовый документ с вложением",
    "description": "Этот документ содержит текстовый файл как вложение",
    "created": "'$(date -I)'",
    "author": "CouchDB Student"
  }')

echo $RESPONSE | python3 -m json.tool

# Извлекаем ID и REV из ответа
DOC_ID=$(echo $RESPONSE | python3 -c "import sys, json; data=json.load(sys.stdin); print(data['id'])")
DOC_REV=$(echo $RESPONSE | python3 -c "import sys, json; data=json.load(sys.stdin); print(data['rev'])")

echo "Создан документ с ID: $DOC_ID и REV: $DOC_REV"
```

```python
#Результат терминала
{
  "ok": true,
  "id": "8f1a2b3c4d5e6f7a8b9c0d1e2",
  "rev": "1-95e6b5c4d3a2b1f0e9d8c7b6a"
}
```

```python
# 3. Создаем текстовый файл для вложения
cat > demo_attachment.txt << 'EOF'

Заголовок: Отчет о выполнении задач CouchDB
Дата: 26 января 2026 г.
Автор: Студент курса баз данных

Задачи выполнены:
1. ✓ Создание документа с определенным _id
2. ✓ Работа с базами данных (создание/удаление)
3. ✓ Работа с вложениями в документах

Примечания:
- Все задачи выполнены через cURL
- Вложения поддерживают различные форматы файлов
- CouchDB хранит метаданные вместе с содержимым

Контакты:
- Email: arevelbakan@gmail.com
- Телефон: +7 (988) 456-78-88

КОНЕЦ ДОКУМЕНТА
EOF

echo "Содержимое файла создано:"
cat demo_attachment.txt
```

```python
# 4. Добавляем текстовый файл как вложение к документу
echo "Добавляем файл как вложение к документу..."
curl -X PUT "http://localhost:5984/attachments_demo/$DOC_ID/report.txt?rev=$DOC_REV" \
  -H "Content-Type: text/plain" \
  --data-binary @demo_attachment.txt
```
